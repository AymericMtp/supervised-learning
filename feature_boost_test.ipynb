```python
import pandas as pd
income_df = pd.read_csv("income_cleaned.csv")
```

```python
income_df['estimation_carriere_age'] = income_df['age'] - 10 - income_df['education-num']
income_df['is_married'] = income_df['marital-status'].apply(lambda x: 1 if 'Married' in x else 0)
income_df.sample(25)
```

```python
numerical_features = ['age', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week', 'estimation_carriere_age', 'is_married']
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))
plt.figure(figsize=(10, 8))
correlation_matrix = income_df[numerical_features + ['income']].corr()

sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation Matrix: Numerical Features + Income', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('correlation_matrix.png', dpi=300, bbox_inches='tight')
plt.show()
```
